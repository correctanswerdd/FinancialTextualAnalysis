```python
# read stopwords file
stopwords = []

f = open("/content/drive/My Drive/Colab Notebooks/text/Stopwords_Bank.txt")
for line in f:
    sw = line.replace('\n', '').replace('\r', '')
    if sw and sw != " ":
        stopwords.append(sw)

# check if there is "blank space" in stopwords
for sw in stopwords:
   if sw == " ":
      print("blank space!")

print(stopwords)
```

```python
# select all characters before the stopword

def clip_text(text, stopwords):
    for sw in stopwords:
        if (sw in text):
            # return text[:re.search(sw, text).start()]
            return text.split(sw)[0], sw
    return text, "STOPWORDS NOT FOUND!"
```

```python
names = []
sw_list = []
n = len(sample.axes[0])
for row in range(n):
    text_content = sample['TextContent'][row]

    # select all characters after the stopword "Primary Bank" && delete"\n""\r"
    prim_bank_and_after = text_content.split('Primary Bank')[1].replace('\n', '').replace('\r', '')

    # select all characters before the stopwords list
    prim_bank, sw = clip_text(prim_bank_and_after, stopwords)
    sw_list.append(sw)

    # to check if there omit some stopwords ... seems not!
    if prim_bank:
        name = prim_bank
    else:
        name = prim_bank_and_after

    # clean the bank name
    name = name.replace('- ', '').replace('-', '')

    # add to list
    names.append(name)

print(names)
# print(len(names))
# print(len(sw_list))
```

```python
df = pd.DataFrame(names) 
df.to_excel (r'/content/drive/My Drive/Colab Notebooks/text/collected_bank_names.xlsx', index = None, header=True)
df = pd.DataFrame(sw_list) 
df.to_excel (r'/content/drive/My Drive/Colab Notebooks/text/used_stopwords.xlsx', index = None, header=True)
```
